```python
## importing the library 
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## creating the url, ticker and API Key
base_url = "https://financialmodelingprep.com/api"
ticker = "TANLA.NS"
api_key = "725362e2869dd5d9f67bc26eb493aefb"
```

```python
## Fetching the TTM data
url = f"{base_url}/v3/ratios-ttm/{ticker}?apikey={api_key}"
response = requests.get(url)
data = response.json()
ratio = pd.DataFrame(data)
ratio
```

```python
##  Fetching the Historic Data 
url1 = f"{base_url}/v3/ratios/{ticker}?limit=40&apikey={api_key}"
response1 = requests.get(url1)
data1 = response1.json()
hist_ratio = pd.DataFrame(data1)
hist_ratio
```

```python
## Dropping symbol and period column
hist_ratio = hist_ratio.drop(columns = ['symbol','period'])
```

```python
## Sorting the values from old to New
hist_ratio = hist_ratio.sort_values(by = 'date', ascending = True)
hist_ratio
```

```python
## Knowing the data type
hist_ratio.info()
```

```python
## Selecting the ratio's which we want 
hist_ratio = hist_ratio[['date','currentRatio','quickRatio','cashRatio','netProfitMargin','returnOnCapitalEmployed','debtEquityRatio','interestCoverage','cashConversionCycle']]
hist_ratio['date'] = pd.to_datetime(hist_ratio['date'])
hist_ratio['year'] = hist_ratio['date'].apply(lambda x: x.year)
hist_ratio
```

```python
## Creating a data frame without date and year for charting purpose
hist_ratio_chart = hist_ratio.drop(columns = ['date','year'])
```

```python
hist_ratio.columns
```

```python
## Creating a chart
sns.set(style="whitegrid", rc={"lines.linewidth": 6, "axes.labelsize": 36, "xtick.labelsize": 34, "ytick.labelsize": 34})
fig, axes = plt.subplots(ncols=1, nrows=len(hist_ratio_chart.columns), figsize=(40, 400)) 
for i, col in enumerate(hist_ratio_chart.columns):
    sns.lineplot(x=hist_ratio['year'], y=hist_ratio_chart[col], ax = axes[i])
    axes[i].set_xlabel(col)
    plt.tight_layout()
```

```python
## Ratios which should be higher
hist_ratio_high = hist_ratio[['date','currentRatio','quickRatio','cashRatio','netProfitMargin','returnOnCapitalEmployed','interestCoverage','cashConversionCycle']]
```

```python
## Ratios which should be lower 
hist_ratio_low = hist_ratio[['date','debtEquityRatio']]
```

```python
## Buy Sell Suggestion 
for ratio in hist_ratio_high.select_dtypes(include=['float64']).columns:
    rolling_mean = hist_ratio_high[ratio].rolling(5).mean()
    hist_ratio_high[f'{ratio}_signal'] = ' '  # Initialize the signal column
    
    # Use .loc to set the 'Buy' and 'Sell' signals
    hist_ratio_high.loc[hist_ratio_high[ratio] > (rolling_mean * 1.2), f'{ratio}_signal'] = 'Buy'
    hist_ratio_high.loc[hist_ratio_high[ratio] < (rolling_mean * 0.8), f'{ratio}_signal'] = 'Sell'
hist_ratio_high
```

```python
for ratio in hist_ratio_low.select_dtypes(include=['float64']).columns:
    rolling_mean = hist_ratio_low[ratio].rolling(5).mean()
    hist_ratio_low[f'{ratio}_signal'] = ' '  # Initialize the signal column
    
    # Use .loc to set the 'Buy' and 'Sell' signals
    hist_ratio_low.loc[hist_ratio_low[ratio] < (rolling_mean * 0.8), f'{ratio}_signal'] = 'Buy'
    hist_ratio_low.loc[hist_ratio_low[ratio] > (rolling_mean * 1.8), f'{ratio}_signal'] = 'Sell'
hist_ratio_low
```
