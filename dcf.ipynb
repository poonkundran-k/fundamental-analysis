```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
base_url = "https://financialmodelingprep.com/api"
ticker = "WSTCSTPAPR.NS"
api_key = "725362e2869dd5d9f67bc26eb493aefb"
```

```python
url_is = f"{base_url}/v3/income-statement/{ticker}?period=annual?limit=120&apikey={api_key}"
response_is = requests.get(url_is)
IS = response_is.json()
IS = pd.DataFrame(IS)
IS['date'] = pd.to_datetime(IS['date'])
IS['year'] = IS['date'].dt.year
IS.sort_values(by = 'year', ascending=True, inplace=True)
IS.set_index(IS['year'], inplace= True)
IS = IS.T
IS
```

```python
url_bs = f"{base_url}/v3/balance-sheet-statement/{ticker}?period=annual?limit=120&apikey={api_key}"
response_bs = requests.get(url_bs)
BS = response_bs.json()
BS = pd.DataFrame(BS)
BS['date'] = pd.to_datetime(BS['date'])
BS['year'] = BS['date'].dt.year
BS.sort_values(by = 'year', ascending=True, inplace=True)
BS.set_index(BS['year'], inplace= True)
BS = BS.T
BS
```

```python
sales = IS.loc['revenue'].astype(float)/10000000
sales
```

```python
sales_growth_rate = np.mean(sales.pct_change())
for year in range(2024, 2029):
    sales[year] = sales[year - 1] * (1 + sales_growth_rate)

sales
sales = sales.loc[2023:]
sales
```

```python
ebitda_margin = IS.loc['ebitda']/IS.loc['revenue']
depr_percent = IS.loc['depreciationAndAmortization']/IS.loc['revenue']
```

```python
ebitda = sales * ebitda_margin.mean()
depreciation = sales * depr_percent.mean()
ebit = ebitda - depreciation
```

```python
nwc_hist = BS.loc['totalCurrentAssets'] - BS.loc['totalCurrentLiabilities']
nwc_percent = nwc_hist / IS.loc['revenue']
```

```python
nwc_percent
```

```python
nwc = sales * nwc_percent.mean()
change_in_nwc = nwc.shift(1) - nwc
```

```python
capex_percent = depr_percent.mean()
capex = -(sales * capex_percent)
```

```python
tax_rate = 0.25
```

```python
tax_payment = -ebit * tax_rate
```

```python
tax_payment = tax_payment.apply(lambda x: min(x, 0))
```

```python
free_cash_flow = ebit + depreciation + tax_payment + capex + change_in_nwc
```

```python
free_cash_flow
```

```python
cost_of_capital = 0.12
terminal_growth = 0.02
```

```python
terminal_value = ((free_cash_flow[2028] * (1 + terminal_growth)) /
                 (cost_of_capital - terminal_growth))
```

```python
discount_factors = [(1 / (1 + cost_of_capital)) ** i for i in range(1,6)]
```

```python
free_cash_flow.loc[2023:]
```

```python
dcf_value = (sum(free_cash_flow.loc[2024:] * discount_factors) +
            terminal_value * discount_factors[-1])
```

```python
output = pd.DataFrame([sales.loc[2023:], ebit.loc[2023:], free_cash_flow.loc[2023:]],
                     index=['Sales', 'EBIT', 'Free Cash Flow']).round(1)
output
```
